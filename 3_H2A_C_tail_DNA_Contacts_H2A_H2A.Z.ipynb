```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib as mpl
```

```python
mpl.rcParams['font.family'] = 'Liberation Sans'
data1 = pd.read_csv("h2a_c_2_Ctail_.out", delim_whitespace=True, header=None)
data2 = pd.read_csv("h2z_c_2_Ctail_.out", delim_whitespace=True, header=None)

time1 = data1[0]  
contacts1 = data1[1]  

time2 = data2[0]  
contacts2 = data2[1]  

window_size = 20  
smooth_contacts1 = contacts1.rolling(window=window_size, center=True).mean()
smooth_contacts2 = contacts2.rolling(window=window_size, center=True).mean()

# Plot
plt.figure(figsize=(10, 6))
plt.plot(time1, smooth_contacts1, color="dimgrey", alpha=0.8, label="H2A H2A_C-tail_1", linewidth=3)
plt.plot(time2, smooth_contacts2, color="darkorange", alpha=0.8, label="H2A.Z H2A.Z_C-tail_1", linewidth=3)
plt.fill_between(time1, contacts1 - np.std(contacts1), contacts1 + np.std(contacts1), color="dimgrey", alpha=0.3)
plt.fill_between(time2, contacts2 - np.std(contacts2), contacts2 + np.std(contacts2), color="darkorange", alpha=0.3)
plt.xlabel("Time ($\mu$s)", fontsize=16, fontweight="bold")
plt.ylabel("No. of contacts", fontsize=16, fontweight="bold")
plt.xticks(fontsize=16, fontweight="bold")
plt.yticks(fontsize=16, fontweight="bold")
plt.legend(loc="upper right", frameon=False, prop={"weight": "bold", "size": 14})

plt.xlim(left=0)  
plt.xlim(0, 1)
plt.ylim(0,170) 
plt.grid(visible=True, linestyle=':', linewidth=0.5, alpha=0.8, color='darkgrey')

plt.gca().spines["top"].set_color("black")
plt.gca().spines["top"].set_linewidth(1.5)
plt.gca().spines["right"].set_color("black")
plt.gca().spines["right"].set_linewidth(1.5)
plt.gca().spines["bottom"].set_color("black")
plt.gca().spines["bottom"].set_linewidth(1.5)
plt.gca().spines["left"].set_color("black")
plt.gca().spines["left"].set_linewidth(1.5)

plt.tight_layout()

plt.savefig("dna_contacts_plot_1.png", dpi=300)
plt.savefig("dna_contacts_plot_1.svg", dpi=300, transparent=True)
plt.show()
```
